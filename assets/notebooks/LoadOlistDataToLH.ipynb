# Transformando os Dados em Tabelas no Lakehouse

```python
from notebookutils import mssparkutils
spark.conf.set("spark.sql.parquet.vorder.enabled", "true")
spark.conf.set("spark.microsoft.delta.optimizeWrite.enabled", "true")
spark.conf.set("spark.microsoft.delta.optimizeWrite.binSize", "1073741824")

csv_directory_path = 'Files/fabric-shortcut-demo/olist-dataset-raw'
# Get list of all CSV files in the directory
# item.path return paths
csv_files = [item.name for item in mssparkutils.fs.ls(csv_directory_path)]

# Loop through each CSV file and save it as a table in the lakehouse
for csv_file in csv_files:
    file_path = f'{csv_directory_path}/{csv_file}'
    table_name = csv_file.split('.')[0]  # Use the file name (without extension) as table name
    schema_name = "bronze"


    # Read the CSV file into a Spark DataFrame
    spark_df = spark.read.format("csv").option("header", "true").load(file_path)
    # Save the Spark DataFrame as a table in the current lakehouse
    # spark_df.write.mode("overwrite").format("delta").save(f"Tables/{table_name}")
    spark_df.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable(f"{schema_name}.{table_name}")
```
